```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from dataset import get_loader
import encoderDecoder
```

```python
val_dataset, val_loader = get_loader(
        root_folder="data/flickr8k/Images",
        captions_file="data/flickr8k/captions_val.txt",
        transform=transform,
        num_workers=2,
    )

train_dataset, train_loader = get_loader(
        root_folder="data/flickr8k/Images",
        captions_file="data/flickr8k/captions_train.txt",
        transform=transform,
        num_workers=2,
    )
print("dataset: " ,train_dataset) 
print("train_loader: " , train_loader)
for imgs, captions in val_loader:
    print(imgs.shape)
    print(captions.shape)
    break
for imgs, captions in train_loader:
    print(imgs.shape)
    print(captions.shape)
    break
```

```python
device = "cuda" if torch.cuda.is_available() else \
         ("mps" if torch.backends.mps.is_available() else "cpu" ) 
print(device) 
```

```python
 train_CNN = False

# Hyperparameters
embed_size = 256
hidden_size = 256
vocab_size = len(train_dataset.vocab)
print(vocab_size)
num_layers = 3
learning_rate = 0.001
num_epochs = 10
drop_prob = 0.5
```

```python
model = encoderDecoder.EncoderDecoder(embed_size, hidden_size, vocab_size, num_layers, drop_prob ).to(device)
```

```python
def indices_to_words(indices, vocab):
    decoded_sentences = []
    count = 0 ; 
    for idx_row in indices:
        count += 1 
        sentence = []
    
        for idx in idx_row:
            word = vocab.get_word(idx.item())
            
            if(word == '<start>'):
                continue 
            if word == '<end>':  # Stop decoding if '<end>' token is encountered
                break
            sentence.append(word)
        decoded_sentences.append(sentence)
    print("the loop runs for" , count)
    return decoded_sentences

from torchtext.data.metrics import bleu_score

for image, caption in val_loader:
    image = image.to(device)
    caption = caption.to(device)
    print(caption)
    print(len(val_dataset.vocab))
    ground_truth_captions = indices_to_words(caption, val_dataset.vocab)
    print(ground_truth_captions[0])
    
    print("caption shape", caption.shape)
    output = model(image, caption)
    print("outputshape" , output.shape)
    _, predicted_indices = torch.max(output, dim=2)
    print("predicted indices", predicted_indices.shape)
    predicted_captions = indices_to_words(predicted_indices, val_dataset.vocab)
    print(predicted_captions[0])
    print("lengths" , len(predicted_captions) ," actual", len(ground_truth_captions))
    bleu_score_value = bleu_score(predicted_captions, ground_truth_captions)
    print("blue score", bleu_score_value)    
```
